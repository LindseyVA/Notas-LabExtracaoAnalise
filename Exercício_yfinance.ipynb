```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#escolhi um ticker - nesse caso pesquisei pela empresa Naver, portal de buscas da CorÃ©ia do Sul, que o ticker Ã© 035420.KS- no site finance.yahoo.com (pesquisar no site)
naver = yf.download('035420.KS', start="2023-01-01")
naver
```

```python
sns.set_theme(style='darkgrid')
sns.lineplot(data=naver, x=naver.index, y='Close', label='035420.KS').set(title='NAVER - 035420.KS')
plt.figure(figsize=(20,10))
```

```python
#MÃ©dia mÃ³vel de 7 dias - para alterar o dia Ã© sÃ³ trocar o nÃºmero dentro de rolling()
naver['MA_7'] = naver['Close'].rolling(7).mean()
naver['MA_7']
```

```python
naver[['Close', 'MA_7']].plot(figsize=(20,10))
```
