```python
url = 'https://ecmi.fgv.br/corpo-docente'
```

```python
options = Options()

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
driver.get(url)
time.sleep(3)
html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```
